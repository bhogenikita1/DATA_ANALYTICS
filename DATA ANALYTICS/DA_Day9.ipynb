# Grids

Grids are general types of plots that allow you to map plot types to rows and columns of a grid, this helps you create similar plots separated by features.

```python
iris=sns.load_dataset("iris")
iris
```

```python
iris.describe()
```

```python
iris.head()
```

```python
iris.shape
```

```python
iris.columns
```

```python
iris["species"].unique()
```

# PairGrid

#Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

```python
sns.PairGrid(iris)
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
g=sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# Pairplot

#pairplot is a simpler version of PairGrid (you'll use quite often)

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue="species",palette="rainbow")
```

# FacetGrid

#FacetGrid is the general way to create grids of plots based off of a feature:

```python
g=sns.FacetGrid(tips,col="time",row="smoker")
```

```python
g=sns.FacetGrid(tips,col="time",row="smoker")
g=g.map(plt.hist,"total_bill")
```

```python
g=sns.FacetGrid(tips,col="time",row="sex")
g=g.map(plt.hist,"total_bill")
```

```python
g=sns.FacetGrid(tips,col="time",row="smoker",hue="sex")
g=g.map(plt.scatter,"total_bill","tip").add_legend()
```

```python
g=sns.FacetGrid(tips,col="day",row="smoker",hue="sex")
g=g.map(plt.scatter,"total_bill","tip").add_legend()
```

# JointGrid

#JointGrid is the general version for jointplot() type grids, for a quick example:

```python
g=sns.JointGrid(x="total_bill",y="tip",data=tips)
g=g.plot(sns.regplot,sns.histplot)
```

# Regression Plot

```python
sns.lmplot(data=tips,x="total_bill",y="tip",hue="sex")
```

```python
sns.lmplot(data=tips,x="total_bill",y="tip",hue="sex",palette='coolwarm',markers=['o','v'],scatter_kws={'s':50})
```

```python
sns.lmplot(data=tips,x="total_bill",y="tip",hue="sex",palette='Set2',markers=['o','v'],scatter_kws={'s':50})
```

```python
sns.lmplot(data=tips,x="total_bill",y="tip",row='sex',col="time")
```
